```python
from diffractio import np, plt, sp
from diffractio import degrees, mm, um
from diffractio.scalar_masks_XY import Scalar_mask_XY
from diffractio.scalar_sources_XY import Scalar_source_XY
from diffractio.utils_drawing import draw_several_fields
```

```python
x = np.linspace(-.1*mm, .1*mm, 1024)
y = np.linspace(-.1*mm, .1*mm, 1024)

wavelength = 0.5*um
```

```python
u0 = Scalar_source_XY(x,y,wavelength)
u0.plane_wave()
```

```python
t0 = Scalar_mask_XY(x,y,wavelength)
t0.circle(r0=(0,0), radius=0.1*mm)
t0.draw()
```

```python
t1 = (t0*u0).RS(z=1*mm)
t1.draw()
```
